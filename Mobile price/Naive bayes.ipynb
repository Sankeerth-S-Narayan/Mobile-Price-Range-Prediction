# Model

### Naive Bayes

```python
from sklearn.model_selection import train_test_split
x=df1.drop(['price_range'],axis=1)
y=df['price_range']
X_train, X_test, y_train, y_test = train_test_split( x ,y, test_size=0.25, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train.ravel())
```

```python
price_predict = model.predict(X_train)
from sklearn import metrics
print("Model Accuracy: {0:.4f}".format(metrics.accuracy_score(y_train, price_predict)))
print()
```

```python
price_test_predict = model.predict(X_test)
print("Model Accuracy: {0:.4f}".format(metrics.accuracy_score(y_test, price_test_predict)))
print()
```

#### Result

Training Model Accuracy: 0.8140

Testing Model Accuracy: 0.8420
